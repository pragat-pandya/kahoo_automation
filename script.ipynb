```python
import pandas as pd
```

```python
cashfree_master = '/home/pragat/Documents/romil_script/cashfree_master.xlsx'
```

```python
cashfree_sheet = pd.read_excel(cashfree_master, skiprows=12)
```

```python
cashfree_sheet = cashfree_sheet.loc[:, ~cashfree_sheet.columns.str.contains('^Unnamed')]
```

```python
partner_groups = cashfree_sheet.groupby('Partner Name')
for partner, data in partner_groups:
    file_name = f"{partner.replace(' ', '_').replace('/', '_')}_bounced_installments.xlsx"
    
    data.to_excel(file_name, index=False)
    
    print(f"Created file: {file_name}")
```
